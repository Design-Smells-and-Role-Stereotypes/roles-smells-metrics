```python
import os
from os.path import exists
import numpy as np
import pandas as pd
```

```python
# create the list of selected projects 

path  = '../../data/Metrics/ck/Mobile/'
projects = [projects for projects in os.listdir(path) if not projects.startswith('.')]
projects
```

```python
h = "project,loc,archSmellCount,dsgnSmellCount,implSmellCount,qualityScore".split(",")
```

```python
from platform import architecture
import xml.etree.ElementTree as ET

weight = (0.5, 0.3, 0.2)

rows = []
for prj in projects:
  file = f"../../data/Smells/designite/Mobile/{prj}/DesigniteAnalysis.xml"
  if exists(file):
    root = ET.parse(file)
    loc = int(root.find(".//Solution/LOC").text)
    arch = int(root.find(".//Solution/TotalArchSmellCount").text)
    dsgn = int(root.find(".//Solution/TotalDesignSmellCount").text)
    impl = int(root.find(".//Solution/TotalImplSmellCount").text)
    dens = tuple((1000/loc)*x for x in (arch,dsgn,impl))
    score = np.dot(weight,dens)
    rows.append([prj,loc,arch,dsgn,impl,score])

df = pd.DataFrame(rows, columns=h)

df.sort_values('qualityScore')
```
